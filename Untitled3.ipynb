```python

import  json

def main(event, context):
    """
   event['body'] is a string dict with the following keys:
   node, event, user, entities.
   Currently, we pass user_id, user_name, full_name, device_platform and language_code in the user dictionary.
   Args:
       event (dict): Data corresponding to this event
       context
   Returns
       (dict): response with statusCode and the response for the User
   """
    body = json.loads(event['body'])
    entities = body.get('entities')
    # ---------------optional--------------------
    entity = Entities(entities)
    user_data = body.get('user')
    conversation_details = body.get('conversation_details')
    final_response = {
        'status': True,
        'entities':entities,
        'user_full_name': user_data.get("full_name"),
        'user_device_platform': user_data.get("device_platform"),
        'conversation_details':conversation_details
    }
    response = {'statusCode': 200, 'body': json.dumps(final_response), 'headers': {'Content-Type': 'application/json'}}
    return response
```

For Fetching entity

```python
class Entities:
    def __init__(self, entities):
        """
        Utility function for entities
        Args:
            entities (dict): The entity object in the mogambo request.
        """
        self.entities = entities

    def get(self, key, default=''):
        """
        Fetches the requested entity
        Args:
            key (str): The name of the entity
            default (str) : What to return if the entity is not found

        Returns:

        """
        entity = self.entities.get(key)
        if not entity:
            return default
        entity = entity[0].get('entity_value', default)
        if entity is not None and 'value' in entity:
            entity = entity['value']
        return entity
```

```python
import base64
from PIL import Image
from io import BytesIO
import time
import requests
import sys  # Import sys module for exiting the script

def base64_to_image(base64_data):
    padding = '=' * (4 - len(base64_data) % 4)
    base64_data_padded = base64_data + padding
    image_data = base64.b64decode(base64_data_padded)
    image_stream = BytesIO(image_data)
    image = Image.open(image_stream)
    return image

def image_to_base64(image, default_format="JPEG"):
    output_stream = BytesIO()

    # Ensure the format is not empty
    format_to_use = image.format if image.format else default_format

    try:
        # Convert the image to 'RGB' mode if the mode is not compatible with JPEG
        if format_to_use.upper() == "JPEG" and image.mode not in ["RGB", "RGBA"]:
            image = image.convert("RGB")

        image.save(output_stream, format=format_to_use)
    except Exception as e:
        raise ValueError(f"Error saving image: {e}")

    base64_data = base64.b64encode(output_stream.getvalue()).decode("utf-8")
    return base64_data

def base64_to_hex(base64_data):
    decoded_data = base64.b64decode(base64_data)
    hex_representation = decoded_data.hex()
    return hex_representation

def resize_image(image, target_size_kb, max_iterations=30):
    target_size_bytes = target_size_kb * 1024

    start_time = time.time()

    for iteration in range(1, max_iterations + 1):
        output_stream = BytesIO()
        try:
            # Ensure the format is not empty
            format_to_use = image.format if image.format else "JPEG"

            # Convert the image to 'RGB' mode if the mode is not compatible with JPEG
            if format_to_use.upper() == "JPEG" and image.mode not in ["RGB", "RGBA"]:
                image = image.convert("RGB")

            image.save(output_stream, format=format_to_use)
        except Exception as e:
            raise ValueError(f"Error saving image: {e}")

        if target_size_bytes <= output_stream.tell() <= 12 * 1024:
            break

        # Calculate the resizing factor based on the original image size
        original_size_kb = output_stream.tell() / 1024
        print("original_size_kb==",original_size_kb)
        if 100 <= original_size_kb < 200:
            print("0.2")
            resizing_factor = 0.2
        elif 200 <= original_size_kb < 1024:
            print("0.1")
            resizing_factor = 0.1
        elif 1024 <= original_size_kb < 2048:
            print("0.1")
            resizing_factor = 0.01
        else:
            resizing_factor = 0.5  # Set a default resizing factor if not in specified ranges

        if output_stream.tell() < target_size_bytes:
            # Resize the image to make it larger
            new_width = max(int(image.width * 2.5), 1)
            new_height = max(int(image.height * 2.5), 1)
        else:
            # Resize the image to make it smaller
            new_width = max(int(image.width * resizing_factor), 1)
            new_height = max(int(image.height * resizing_factor), 1)

        image = image.resize((new_width, new_height))

    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Resizing took {iteration} iterations and {elapsed_time:.2f} seconds")

    return image



def image_url_to_base64(image_url):
    try:
        # Download the image
        response = requests.get(image_url, stream=True)
        response.raise_for_status()

        # Convert the image to base64
        base64_data = base64.b64encode(response.content).decode("utf-8")

        return base64_data

    except Exception as e:
        print(f"Error converting image URL to base64: {e}")
        return None

image_url = "https://media.istockphoto.com/id/1494104649/photo/ai-chatbot-artificial-intelligence-digital-concept.jpg?s=1024x1024&w=is&k=20&c=FQZBVxbXjhgPzm-W-5Jo3O7x7OXLPugwXzTKEu6nV3w="

base64_data = image_url_to_base64(image_url)

if base64_data:
    print("Base64 data:", base64_data)

# Check the original image size
original_size_bytes = len(base64_data)
print("original_size_bytes---",original_size_bytes)
original_size_kb = original_size_bytes / 1024
print(f"Original image size: {original_size_kb:.2f} KB")

# # Additional condition to check if the original image size is less than 1 KB
# if original_size_bytes < 1024:
#     print("Image size is less than 1 KB. Exiting.")
#     # Convert base64 to hex
#     hex_code = base64_to_hex(base64_data)
#     print("Hex code:", hex_code)
#     sys.exit(1)

if original_size_bytes > 1.2 * 1024 * 1024:  # Check if greater than 1 MB
    print("Image size is greater than 1 MB. Exiting.")
    # Convert base64 to hex
    hex_code = base64_to_hex(base64_data)
    print("Hex code:", hex_code)
    sys.exit(1)


# Convert base64 to image
original_image = base64_to_image(base64_data)

# Resize the image
resized_image = resize_image(original_image, target_size_kb=4, max_iterations=30)

# Check the resized image size
resized_size_kb = len(image_to_base64(resized_image)) / 1024
print(f"Resized image size: {resized_size_kb:.2f} KB")

# Convert the final resized image to base64
base64_resized = image_to_base64(resized_image)
print("base64_resized--",base64_resized)

# Convert base64 to hex
hex_code = base64_to_hex(base64_resized)

# Print the final result
print(f"Final Resized image size: {resized_size_kb:.2f} KB")
print("Hex code:", hex_code)
```
